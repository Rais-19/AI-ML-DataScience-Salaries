The data reveals that medium-sized companies (M) offer higher average salaries (152K) than large companies (L) (135K), despite expectations that larger firms would pay more. This counterintuitive trend can be attributed to two key factors:

(2) medium-sized companies—often fast-growing tech firms or specialized players—may aggressively compete for top talent by offering premium salaries, while large corporations balance compensation across broader, more standardized pay bands.

```python
def get_country_name(code):
    try:
        # Handle special cases first
        if code == 'PR':
            return 'Puerto Rico'
        elif code == 'US':
            return 'United States'
        elif code=='QA':
            return 'Qatar'
        elif code=='CA':
            return 'Canada'
        elif code=='IL':
            return 'Israel'
        elif code=='AU':
            return 'Australia'
        elif code=='SA':
            return 'Saudi Arabia'
        elif code=='CZ':
            return 'Czech Republic'
        elif code == "VE":
          return "Venezula"
        elif code=='EG':
            return 'Egypt'
        # Use pycountry for other codes
        country = pycountry.countries.get(alpha_2=code)
        return country.name if country else code
    except:
        return code

# Apply the function to create a new column with full names
top_10_loc['country_name'] = top_10_loc['company_location'].apply(get_country_name)

sns.barplot(x=top_10_loc['salary_in_usd'], y=top_10_loc['country_name'], palette="viridis")
```

```python
df['work_type'] = df['remote_ratio'].apply(
    lambda x: 'Remote' if x == 100 else ('On-site' if x == 0 else 'Hybrid')
)
remote_salary = df.groupby('work_type')['salary_in_usd'].mean().sort_values(ascending=False)
print(remote_salary)
```

Key Findings:

On-Site Dominance - Highest average pay at $153K, reflecting traditional tech/finance hubs where physical presence commands premium compensation.

Remote Close Second - $147K shows strong viability for distributed work, though slight discount versus on-site suggests location-based pay adjustments.

Hybrid Penalty - Sharp drop to $81K (-47% vs on-site) indicates these roles may be concentrated in lower-cost regions while More junior-level positions are subject to employer cost-saving strategies

```python
sns.barplot(data=df, x='work_type', y='salary_in_usd', estimator='mean', palette='Set2', errorbar=None)
plt.title('Average Salary by Work Type')
plt.xlabel('Work Type')
plt.ylabel('Average Salary (USD)')
plt.show()
```

It is evident that hybrid jobs offer significantly lower salaries compared to on-site and remote positions. Based on AI analysis and insights from various articles, we propose the following three hypotheses to explain this substantial difference:

1:Seniority Bias - Hybrid roles may be disproportionately entry-level/junior positions

2:Geographic Arbitrage - Employers could be placing hybrid jobs in lower-cost regions while maintaining HQ salaries for on-site staff

3:Role-Type Segregation - Hybrid arrangements might cluster in support functions (e.g., data engineering) versus core AI/ML roles

```python
remote_by_role = df.groupby(['job_title', 'work_type']).size().unstack()
remote_by_role = remote_by_role.div(remote_by_role.sum(axis=1), axis=0)
top_roles = df['job_title'].value_counts().head(10).index
remote_by_role = remote_by_role.loc[top_roles]
print(remote_by_role)
```

An analysis of the top 10 highest-paying job titles reveals that on-site roles dominate, consistently accounting for over 70% of positions across titles like Manager, Engineer, and Data Scientist.
Remote work follows, representing roughly 20–38% depending on the role, with Analytics Engineer showing the highest remote ratio at 38%.
Hybrid roles remain marginal, comprising less than 2% across most job titles, suggesting limited availability or lower valuation.

```python
remote_by_role.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Remote Work Type by Job Title (Top 10)')
plt.xlabel('Job Title')
plt.ylabel('Proportion')
plt.legend(title='Work Type', bbox_to_anchor=(1.05, 1))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
remote_by_size = df.groupby(['company_size', 'work_type']).size().unstack()
remote_by_size = remote_by_size.div(remote_by_size.sum(axis=1), axis=0)
print(remote_by_size)
```

Small companies (S) strongly favor remote work, with over 55% of roles being remote — suggesting agility, reduced overhead, and wider talent access.
In contrast, medium-sized companies (M) lean heavily toward on-site roles (75%), indicating structured workflows and in-office collaboration priorities.
Large companies (L) offer the most balanced model, with 15% hybrid, 59% on-site, and 26% remote, reflecting their capability to support flexible work environments.
The sharp contrast in hybrid adoption—low in M (0.1%) and high in S (19.9%) and L (15%)—highlights where flexibility is emerging or still underused.

```python
remote_by_size.loc[['S', 'M', 'L']].plot(
    kind='bar', stacked=True, figsize=(8, 5)
)
plt.title('Remote Work Type by Company Size')
plt.xlabel('Company Size')
plt.ylabel('Proportion')
plt.legend(title='Work Type')
plt.tight_layout()
plt.show()
```

```python
remote_by_exp = df.groupby(['experience_level', 'work_type']).size().unstack()
remote_by_exp = remote_by_exp.div(remote_by_exp.sum(axis=1), axis=0)
print(remote_by_exp)
```

Across all experience levels, on-site roles are the dominant format, representing around 72–79% of positions, with the highest share among entry-level (79%) professionals.

Executives (EX) and senior professionals (SE) are more likely to work remotely than entry-level peers, suggesting higher trust and autonomy.
These trends reflect a clear industry preference for on-site presence early in careers, transitioning toward remote flexibility at higher levels.
